## **Introduzione :**

### Biografia

Billy Beane è un ex giocatore di baseball professionista e un dirigente sportivo noto per il suo rivoluzionario approccio analitico al baseball, noto come "Moneyball". Nato il 29 marzo 1962 a Orlando, Florida, Beane ha avuto una carriera da giocatore di baseball prima di diventare un dirigente di successo.

Beane è stato selezionato come 23º pick assoluto nel draft del Major League Baseball (MLB) del 1980 dai New York Mets. Ha fatto il suo debutto nella MLB nel 1984 con i Mets, ma è stato scambiato con gli Oakland Athletics nel 1985. La sua carriera da giocatore non è stata eccezionale, ma ha guadagnato una reputazione come un ragazzo con un potenziale incredibile.

Dopo il ritiro dal gioco attivo nel 1990, Beane è entrato nella gestione dei giocatori degli Oakland Athletics. Tuttavia, è stato il suo ruolo come general manager degli A's che ha portato alla creazione del famoso algoritmo Moneyball. L'idea di Moneyball è nata dalla necessità di trovare un modo per competere con squadre che avevano budget molto più grandi per la stipula di contratti di giocatori di alto livello.

#### Moneyball

L'ispirazione principale per Moneyball è venuta da un giovane analista del baseball di nome Paul DePodesta, che ha portato Beane a considerare l'uso di statistiche avanzate e analisi dettagliate per valutare i giocatori in modo diverso. Invece di concentrarsi solo su statistiche tradizionali come la media battuta, fuoricampo o punti battuti a casa, Beane e il suo team hanno iniziato a guardare statistiche più avanzate come l'OPS (On-Base Plus Slugging), l'OBP (On-Base Percentage) e altri indicatori che avevano dimostrato una migliore correlazione con il successo della squadra.

L'obiettivo di Moneyball era identificare giocatori sottovalutati e sottovalutati che potevano essere acquistati a un costo relativamente basso, ma che avevano un impatto significativo sulla squadra. Beane ha utilizzato queste analisi per assemblare una squadra competitiva degli Athletics nonostante il loro budget limitato.

Il libro "Moneyball: The Art of Winning an Unfair Game" scritto da Michael Lewis nel 2003 ha reso famosa questa strategia e ha portato alla sua diffusione nell'intero mondo del baseball. Il libro è stato successivamente adattato in un film omonimo nel 2011, con Brad Pitt nel ruolo di Billy Beane.

L'approccio di Billy Beane e Moneyball ha avuto un impatto duraturo sul baseball professionistico e ha contribuito a cambiare il modo in cui le squadre valutano e scelgono i giocatori. Beane è rimasto un importante dirigente di baseball e un leader nell'uso delle analisi avanzate per prendere decisioni nel mondo dello sport.

## **Come funziona il metodo Moneyball?**

**Selezione delle Statistiche Chiave**: Nel metodo Moneyball, il processo inizia con l'identificazione delle statistiche più significative per valutare le prestazioni di un giocatore. Queste statistiche avanzate sostituiscono spesso le tradizionali, come la media battuta o i fuoricampo, perché offrono una visione più accurata delle abilità di un giocatore.
Questo sistema analizza e incrocia i dati dei vari atleti: non solo le statistiche e le caratteristiche fisiche dei giocatori, ma anche la possibilità di adattamento in un determinato contesto tattico e ambientale.

**Identificazione dei Giocatori Sottovalutati**: Un elemento fondamentale del Moneyball è individuare giocatori che sono spesso trascurati dal mercato, ma che hanno potenzialità significative. Questi giocatori possono essere meno conosciuti o non avere contratti costosi, ma le loro statistiche avanzate mostrano il loro valore reale. L'analisi avanzata permette di scovare queste "hidden gems" nel mondo del baseball.

**Costruzione del Roster**: Gli Oakland Athletics, sotto la guida di Billy Beane, utilizzano queste analisi per selezionare i giocatori del loro roster. Si concentrano su atleti che, in base alle statistiche chiave, offrono un buon rapporto tra il loro valore e il loro costo. Questo approccio consente loro di massimizzare le risorse disponibili, ottimizzando il valore del roster. La squadra cerca di bilanciare giocatori che eccellono in diverse abilità, come il lancio, la difesa e il corridoio delle basi.

**Flessibilità Tattica**: Moneyball non riguarda solo la selezione dei giocatori, ma anche l'adozione di una strategia di gioco basata su dati. Gli Athletics cercano di ottenere basi attraverso il lancio paziente, l'abilità di raggiungere le basi e una solida difesa. Questa strategia mira a sfruttare al massimo le abilità dei giocatori selezionati e a creare opportunità per segnare punti e vincere partite.

**Risultati**: Nonostante il budget limitato, gli Athletics hanno ottenuto successo con l'approccio Moneyball, raggiungendo i playoff in più occasioni. Questo ha dimostrato che è possibile competere con squadre con risorse finanziarie molto più elevate utilizzando analisi avanzate e strategie basate sui dati.


#**La Sabermetrica**


Il baseball è popolare in diversi paesi, e negli ultimi decenni, l'analisi statistica nel baseball, nota come sabermetrica, ha trasformato il modo in cui giocatori e squadre vengono valutati. Questa rivoluzione è stata avviata da Bill James negli anni '70, un appassionato senza esperienza professionale nel baseball. Ha introdotto il termine "sabermetrics" per descrivere l'analisi dettagliata del gioco basata su dati di prestazioni anziché sulle statistiche tradizionali come la media battuta.

Le statistiche sabermetriche utilizzano misure più complesse. Queste statistiche più dettagliate forniscono una visione più accurata delle prestazioni di un giocatore rispetto alle statistiche tradizionali. Nonostante le statistiche sabermetriche non siano comunemente presenti nei tabellini di gioco, i loro calcoli sono utilizzati dagli esperti per valutare i giocatori in modo più approfondito. Questa nuova prospettiva ha contribuito a cambiare la percezione del valore dei giocatori e ha portato a una migliore comprensione del gioco del baseball.

Beane e la sua dirigenza focalizzata sulla sabermetrica hanno sostenuto statistiche come la percentuale di arrivo in base (OBP - [colpi più BB più HB] diviso per [turni in battuta più BB più HB più SF]) e la percentuale di slugging (SP - basi totali ottenute dai colpi diviso per il numero di turni in battuta) perché credevano che queste misure "... rappresentino i due componenti essenziali della creazione dell'offesa

BB = Si verifica una base su ball (base on balls, nella dizione americana anche "walk", "camminata", per il fatto che il battitore che la riceve può tranquillamente camminare verso la prima base) quando il lanciatore effettua quattro lanci in zona non valida di strike allo stesso battitore, senza che questi metta in gioco la palla con una battuta o venga eliminato con uno strikeout.

HB = Il "hit by pitch" (HBP) è un evento in cui un battitore o i suoi vestiti o attrezzature (diverse dalla mazza) sono colpiti direttamente da un lancio del lanciatore.

SF = Un "sacrifice fly" : "Segna un 'sacrifice fly' quando, prima che siano fatte due eliminazioni, il battitore colpisce una palla in volo gestita da un esterno o da un interno che corre nel campo esterno, sia in territorio valido che in territorio non valido, e

*   viene presa, e una corsa viene segnata dopo la presa
*   viene lasciata cadere, e un corridore segna, se secondo il giudizio del segnapunti il corridore avrebbe potuto segnare dopo la presa se la palla fosse stata presa."

Una statistica tradizionale che viene spesso utilizzata per valutare giocatori e squadre è quella delle basi rubate. Tuttavia, i sabermetrici
sostengono che i numeri base rubati distorcono il valore di un giocatore.

Il collaboratore di Baseball Prospectus Clay Davenport ha sviluppato la statistica di Fielding Runs Above Average
(FRAA), che si adatta a fattori come il parco da gioco (tutti i parchi differiscono per dimensioni e forma), il totale delle palle in
gioco consentite, le tendenze del groundball o della palla volante.
Il numero di un difensore rappresenta quanti punti ha salvato alla squadra
o costato alla squadra rispetto a un difensore "medio" calcolato per il campionato

UZR utilizza i dati registrati partita su partita da Baseball Info Solutions negli ultimi anni per dare credito in campo a giocatori specifici. Il "motore"
dell'UZR è in grado di compilare il tipo, la posizione e la velocità di ciascuna palla battuta e anche di prendere in considerazione dati come la
temperatura e lo stadio in cui si gioca.

La terza e ultima area che l’analisi sabermetrica può (presumibilmente) aiutare ad analizzare meglio è la prestazione del lancio.
Recentemente, molti appassionati di baseball, soprattutto quelli che partecipano a leghe di fantasy baseball, hanno cominciato a utilizzare statistiche come il WHIP (valore delle basi per inning lanciato) e il K/9 (strikeout per nove inning) per valutare adeguatamente il valore di un lanciatore. Il WHIP e il K/9 permettono di valutare quanto bene ogni lanciatore si comporti realmente sul monte di lancio, senza considerare il contesto. Questi dati risultano più affidabili rispetto a statistiche come le vittorie, che sono troppo legate al contesto di gioco. Infatti, un lanciatore eccellente non dovrebbe essere sottovalutato solo perché fa parte di una squadra con una scarsa capacità offensiva (cioè, le stesse prestazioni possono portare a meno vittorie).

Il seguente codice potrebbe essere una rappresentazione simile di dati presi in input dal team di Owramipur per poi usare una rete "bayesiana".

```python
import pandas as pd
import numpy as np

# Creo un DataFrame

squadre = ['Barcellona', 'Avversari']
num_squadre = len(squadre)

dati = {
    'Squadra': squadre,
    'Motivazione': np.random.randint(1, 11, num_squadre),
    'Concentrazione': np.random.randint(1, 11, num_squadre),
    'Determinazione': np.random.randint(1, 11, num_squadre),
    'Grinta': np.random.randint(1, 11, num_squadre),
    'ResistenzaMentale': np.random.randint(1, 11, num_squadre),
    'Pressione': np.random.randint(1, 11, num_squadre),
    'PossessoPalla': np.random.randint(40, 71, num_squadre),
    'MediaTiriInPorta': np.random.randint(5, 16, num_squadre),
    'MediaGoalSegnati': np.random.randint(0, 4, num_squadre),
    'MediaGoalSubiti': np.random.randint(0, 4, num_squadre),
    'PosizioneClassifica': np.random.randint(1, 21, num_squadre),
}

# Calcola le vittorie, i pareggi e le sconfitte nelle ultime 5 partite
dati['VittorieUltimi5'] = np.random.randint(0, 6, num_squadre)
dati['PareggiUltimi5'] = np.random.randint(0, 6 - dati['VittorieUltimi5'], num_squadre)
dati['SconfitteUltimi5'] = 5 - dati['VittorieUltimi5'] - dati['PareggiUltimi5']

# Calcola i punti in base ai risultati delle partite
dati['PuntiUltimi5'] = dati['VittorieUltimi5'] * 3 + dati['PareggiUltimi5']

# Crea il DataFrame
df = pd.DataFrame(dati)

# Mostra il DataFrame
for index, row in df.iterrows():
    print(f"{row['Squadra']} - Vittorie: {row['VittorieUltimi5']}, Pareggi: {row['PareggiUltimi5']}, Sconfitte: {row['SconfitteUltimi5']}")
print(df)
```

###**Cos'è una reta "bayesiana"?  (BN)**

Una rete bayesiana, o rete bayesiana probabilistica, è un modello grafico probabilistico che rappresenta le relazioni probabilistiche tra un insieme di variabili casuali. Queste reti sono chiamate "bayesiane" perché sono basate sul teorema di Bayes, che descrive come le probabilità delle cause possono essere aggiornate in base alle osservazioni.
Le reti bayesiane vengono utilizzate per modellare e risolvere problemi di ragionamento probabilistico e di previsione. Le reti bayesiane sono particolarmente utili quando ci sono incertezze nel sistema o quando è necessario combinare conoscenze soggettive con dati osservabili per prendere decisioni. Sono utilizzate in diversi campi, come intelligenza artificiale, medicina, finanza e ingegneria.

##Vocabolario delle metriche

Accuracy (Accuratezza):

L'accuracy è la proporzione di previsioni corrette rispetto al totale delle previsioni.
Formula: Accuracy = (TP + TN) / (TP + TN + FP + FN)
Rappresenta la percentuale di correttezza globale del modello.
Precision (Precisione):

La precision è la proporzione di istanze predette come positive che sono effettivamente positive.
Formula: Precision = TP / (TP + FP)
Misura la precisione del modello quando prevede la classe positiva.
Recall (Recupero o Sensibilità):

La recall è la proporzione di istanze positive reali che sono state predette correttamente.
Formula: Recall = TP / (TP + FN)
Misura la capacità del modello di identificare correttamente tutte le istanze positive.
F1-score:

L'F1-score è la media armonica tra precision e recall. È utile quando si desidera bilanciare precision e recall.
Formula: F1-score = 2 * (Precision * Recall) / (Precision + Recall)
Fornisce una singola misura che combina precisione e recall.
ROC AUC (Area sotto la curva ROC):

ROC AUC misura la capacità del modello di distinguere tra le classi positiva e negativa variando la soglia di classificazione.
ROC AUC è l'area sotto la curva ROC, che rappresenta il tasso di vero positivo (sensibilità) rispetto al tasso di falso positivo (1-specificità).
Un valore ROC AUC vicino a 1 indica un buon modello, mentre un valore vicino a 0.5 indica un modello casuale.

```python
from sklearn.ensemble import RandomForestClassifier #random forest algoritmo basato su alberi decisionali
rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1) #Viene creato un oggetto di RandomForestClassifier con i parametri specificati: 50 alberi decisionali (n_estimators=50), una divisione minima di campioni in un nodo interno pari a 10 (min_samples_split=10), e uno stato random fissato a 1 (random_state=1).
train = matches[matches["Date"] < '2023-05-01'] #suddivisione in train e test
test = matches[matches["Date"] > '2023-05-01']
predictors = ["venue_code", "opp_code", "hour", "day_code"] #parametri su cui basiamo la nostra previsione, ovviamente parametri che si sanno prima della partita questo renderà la precisione meno alta
rf.fit(train[predictors], train["target"])
preds = rf.predict(test[predictors])
from sklearn.metrics import accuracy_score
combined = pd.DataFrame(dict(actual=test["target"], predicted=preds))
pd.crosstab(index=combined["actual"], columns=combined["predicted"])
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Calcola la matrice di confusione
conf_matrix = confusion_matrix(test["target"], preds)

plt.figure(figsize=(6, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=["Predicted 0", "Predicted 1"], yticklabels=["Actual 0", "Actual 1"])
plt.title('Confusion Matrix')
plt.show()
```

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("Date")
    rolling_stats = group[cols].rolling(10, closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
cols = ["GF","GA","Sh","SoT","Dist","FK","PK","PKatt","Cmp%","CrsPA","Poss"]
new_cols = [f"{c}_rolling" for c in cols]
```

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("Date")
    rolling_stats = group[cols].rolling(5, closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
cols = ["GF","GA","Sh","SoT","Dist","FK","PK","PKatt","Cmp%","CrsPA","Poss"]
new_cols = [f"{c}_rolling" for c in cols]
matches_rolling = matches_prezzi.groupby("Team").apply(lambda x: rolling_averages(x, cols, new_cols))
matches_rolling = matches_rolling.droplevel('Team')
```
